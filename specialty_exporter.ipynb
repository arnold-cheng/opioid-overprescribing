```python
import pandas as pd
import numpy as np
```

```python
# if your computer can handle large files, you should be able to run this code cell
medicare = pd.read_csv('Medicare_Provider_Utilization_and_Payment_Data__2015_Part_D_Prescriber.csv')
```

```python
# Here is a list of all different specialties that you can select as your 
list(medicare['specialty_description'].unique())
```

```python
def specialty_exporter(data, specialties):
    for specialty in specialties:
        df = data[data['specialty_description'] == specialty]
        file_name = specialty.replace(' ', '_').lower()
        df.to_csv(f'{file_name}.csv', index=False)
```

```python
# example
specialty_exporter(medicare, ['General Surgery'])
```
